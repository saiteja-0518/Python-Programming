Q9. Using this class:

Q11 : Employee Salary Calculator
Create a class Employee with attributes: name, salary, department
Add a method annual_salary() that returns salary × 12
Add a method give_raise(percentage) that increases salary by given percentage
Create 2 employees, display their annual salary, give them a 10% raise, and display new annual salary


```python
class Employee:
    def __init__(self, name, salary, department):
        self.name = name
        self.salary = salary
        self.department = department

    def annual_salary(self):
        return self.salary * 12

    def give_raise(self, percentage):
        self.salary += self.salary * percentage / 100

e1 = Employee("Teja", 55000, "IT")
e2 = Employee("Pavan", 48000, "HR")

print(e1.annual_salary())
e1.give_raise(10)
print(e1.annual_salary())

print(e2.annual_salary())
e2.give_raise(10)
print(e2.annual_salary())
```

Q12 : Rectangle Class
Create a class Rectangle with attributes: length, width
Add methods:
area() → returns length × width
perimeter() → returns 2 × (length + width)
is_square() → returns True if length equals width, else False
Create 3 rectangles and display their area, perimeter, and whether they're squares


```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def is_square(self):
        return self.length == self.width

r1 = Rectangle(5, 5)
r2 = Rectangle(4, 6)
r3 = Rectangle(3, 7)

for r in [r1, r2, r3]:
    print(f"Area: {r.area()}, Perimeter: {r.perimeter()}, Is Square: {r.is_square()}")
```

Q13 : Temperature Converter
Create a class Temperature with attribute: celsius
Add methods:
to_fahrenheit() → returns (celsius × 9/5) + 32
to_kelvin() → returns celsius + 273.15
is_freezing() → returns True if celsius ≤ 0
Create 3 temperature objects and display conversions


```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15

    def is_freezing(self):
        return self.celsius <= 0

t1 = Temperature(0)
t2 = Temperature(25)
t3 = Temperature(-10)

for t in [t1, t2, t3]:
    print(f"Celsius: {t.celsius}, Fahrenheit: {t.to_fahrenheit()}, Kelvin: {t.to_kelvin()}, Freezing: {t.is_freezing()}")
```

Q14 : Movie Rating System
Create a class Movie with attributes: title, director, rating (out of 10)
Add methods:
display_info() → prints movie details
rating_category() → returns "Excellent" if rating ≥ 8, "Good" if ≥ 6, "Average" if ≥ 4, "Poor" if < 4
is_blockbuster() → returns True if rating ≥ 8
Create 4 movie objects with different ratings and display their category and blockbuster status


```python
class Movie:
    def __init__(self, title, director, rating):
        self.title = title
        self.director = director
        self.rating = rating

    def display_info(self):
        print(f"Title: {self.title}, Director: {self.director}, Rating: {self.rating}")

    def rating_category(self):
        if self.rating >= 8:
            return "Excellent"
        elif self.rating >= 6:
            return "Good"
        elif self.rating >= 4:
            return "Average"
        else:
            return "Poor"

    def is_blockbuster(self):
        return self.rating >= 8

m1 = Movie("RRR", "Rajamouli", 9)
m2 = Movie("KGF", "Prashanth", 7)
m3 = Movie("Baahubali", "Rajamouli", 8)
m4 = Movie("Pushpa", "Sukumar", 3)

for m in [m1, m2, m3, m4]:
    m.display_info()
    print(f"Category: {m.rating_category()}, Blockbuster: {m.is_blockbuster()}")
    print()
```
