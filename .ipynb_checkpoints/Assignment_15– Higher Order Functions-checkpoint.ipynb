Q1. Use map() with a function to get the length of each word in a list.
Input: ['apple', 'banana', 'cherry']
Output: [5, 6, 6]


```python
word = ['apple', 'banana', 'cherry']
r = list(map(lambda w: len(w),word))
print(r)
```

Q2. Use filter() with a function to keep only strings longer than 3 characters.
Input: ['hi', 'hello', 'hey', 'world']
Output: ['hello', 'world']


```python
def s(word):
    return len(word)>3
w =  ['hi', 'hello', 'hey', 'world']
a = list(filter(s,w))
print(a)
```

Q3. Use reduce() with a function to find the maximum number in a list.
Input: [10, 25, 7, 99, 34]
Output: 99



```python
from functools import reduce
def maxi(a,b):
    return max(a,b)
num = [10, 25, 7, 99, 34]
print(reduce(maxi,num))
```

Q4. Use filter() with a function to remove empty strings from a list.
Input: ['apple', '', 'banana', '', 'cherry']
Output: ['apple', 'banana', 'cherry']


```python
def t(word):
    return bool(word)
w =['apple', '', 'banana', '', 'cherry']
print(list(filter(t,w)))
```

Q5. Use map() with a function to convert temperatures from Celsius to Fahrenheit.
Input: [0, 25, 100]
Output: [32, 77, 212]


```python
def fahrenheit(c):
    return  int(c*9/5) +32
t = [0,25,100]
r =list(map(fahrenheit,t))
print(r)
```

Q6. Create a lambda to check if a number is positive.
Input: 5
Output: "Positive"


```python
l = lambda m :"Positive" if m>0 else "Not Positive"
print(l(5))
```

Q9. Use reduce() with a lambda to compute the product of all numbers.
 Input: [2, 3, 4, 5]
 Output: 1

```python
from functools import reduce
l = [2, 3, 4, 5]
r = reduce(lambda a,b :a*b,l)
print(r)
```

Q10. Use map() with a lambda to title-case each string.
 Input: ['hello world', 'python programming']
 Output: ['Hello World', 'Python Programming']


```python
k = ['hello world', 'python programming']
r = list(map(lambda x:x.title(), k))
print(r)
```

Q11. Use filter() with a lambda to keep palindromic strings.
 Input: ['level', 'world', 'radar', 'hello', 'civic']
 Output: ['level', 'radar', 'civic']


```python
w = ['level', 'world', 'radar', 'hello', 'civic']
h = list(filter(lambda x:x==x[::-1],w))
print(h)
```

Q12. Use reduce() with a lambda to concatenate a list of strings with a hyphen.
 Input: ['a', 'b', 'c', 'd']
 Output: 'a-b-c-d'


```python
from functools import reduce
l = ['a', 'b', 'c', 'd']
d = reduce(lambda x,y :x+"-"+y,l)
print(d)
```

Q13. Use map() with a lambda to extract domains from emails (text after '@').
 Input: ['alice@mail.com', 'bob@work.org', 'care@health.in']
 Output: ['mail.com', 'work.org', 'health.in']

```python
email = ['alice@mail.com', 'bob@work.org', 'care@health.in'] 
t = list(map(lambda x:x.split("@")[1],email))
print(t)
```

Q14. Use filter() with a lambda to keep numbers that are multiples of both 2 and 3.
 Input: [3, 6, 9, 12, 15, 18, 20]
 Output: [6, 12, 18]


```python
n = [3, 6, 9, 12, 15, 18, 20]
r = list(filter(lambda x:x %2 ==0 and x%3 ==0 ,n))
print(r)
```

Q15. Use reduce() with a lambda to find the longest string (by length).
 Input: ['ant', 'butterfly', 'cat', 'dinosaur']
 Output: 'butterfly'


```python
u = ['ant', 'butterfly', 'cat', 'dinosaur']
r = reduce(lambda x,y :x if len(x) >len(y) else y,u)
print(r)
```
