## Set, Dictionary

Part A: Set

Q1. Find common elements from three sets.
 Input: {1,2,3,4}, {2,3,5,6}, {0,2,3,7}
 Output: {2, 3}




```python
set1 ={1,2,3,4}
set2 ={2,3,5,6}
set3 ={0,2,3,7}
r = set1 & set2 & set3
print("common elements:",r)
```

Q2. Find elements that are only in the first set but not in the other two.
 Input: {1,2,3,4,5}, {2,3}, {4,6}
 Output: {1, 5}






```python
s1={1,2,3,4,5}
s2 ={2,3}
s3 ={4,6}
s = s1-s2-s3
print(s)
```

Q3. Create a set of unique vowels from a string.
 Input: "education"
 Output: {'a', 'e', 'i', 'o', 'u'}

```python
i="education" 
j = {'a','e','i','o','u'}
print(set(i)&j)
```

Q4. Convert a list of tuples into a set.
 Input: [(1,2), (3,4), (1,2)]
 Output: {(1, 2), (3, 4)}


```python
o =[(1,2), (3,4), (1,2)] 
p =set(o)
print(p)
```

Q5. Find the intersection of characters in two words.
 Input: "banana", "bandana"
 Output: {'a', 'b', 'n'}



```python
a ="banana"
b ="bandana"
s = set(a)
z = set(b)
print(s&z)
```

Q6. Find digits in the first number but not in the second.
 Input: 12345, 34567
 Output: {'1', '2'}


```python
e =12345
t =34567
a =set(str(e))
b =set(str(t))
print(a-b)
```

Q7. Find all unique words from a sentence.
 Input: "python is fun and python is powerful"
 Output: {'is', 'python', 'fun', 'and', 'powerful'}



```python
t ="python is fun and python is powerful"
w = t.split()
i =set(w)
print(i)
```

Q8. Remove duplicates from a list.
 Input: [1,2,2,3,3,4,5]
 Output: {1, 2, 3, 4, 5}

```python
p=[1,2,2,3,3,4,5]
print(set(p))
```

Q9. Find non-overlapping characters between two words.
 Input: "hello", "world"
 Output: {'h', 'e', ‘w’, 'r', 'd'}


```python
o =set("hello")
p=set("world")
print(o ^ p)
```

Q10. Find maximum and minimum elements in a set.
 Input: {12, 7, 25, 3, 18}
 Output: 25 3


```python
w ={12, 7, 25, 3, 18}
ma =max(w)
mi =min(w)
print("Max:",ma)
print("Min",mi)
```

Part B: Dictionary

Q1. Get the value of a dictionary with the maximum value.
 Input: {'a':10, 'b':25, 'c':15}
 Output: 25


```python
o ={'a':10, 'b':25, 'c':15}
k =max(o.values())
print(k)
```

Q2. Get all dictionary values as a single string.
 Input: {'a':'hi','b':'there'}
 Output: "hi there"



```python
l ={'a':'hi','b':'there'} 
print(l['a']+" "+l['b'])
```

Q3. Remove the last inserted pair.
Input: {'a':1, 'b':2, 'c':3}
Output: {'a':1, 'b':2}




```python
o ={'a':1, 'b':2, 'c':3}
o.popitem()
print(o)
```

Q4. Find values present in both dictionaries.
 Input: {'x':1, 'y':2}, {'a':2, 'b':3}
 Output: {2}


```python
i ={'x':1, 'y':2}
j ={'a':2, 'b':3}
print(set(i.values()) & set(j.values()))
```

Q5. Find the sum of dictionary keys.
 Input: {10:'a', 20:'b', 30:'c'}
 Output: 60


```python
d ={10:'a', 20:'b', 30:'c'}
p =sum(d.keys())
print(p)
```

Q6. Remove duplicates from dictionary values.
 Input: {'a':1, 'b':2, 'c':1}
 Output: {1, 2}


```python
l={'a':1, 'b':2, 'c':1}
print(set(l.values()))
```
