# Nested Dictionary

Create the dictionary and answer the following questions:


```python
mobile = {
    'Apple': {
        '13': [38000, 45000, 52000],
        '14': [56000, 60000, 70000],
        '15': [67000, 72000, 82000]
    },


    'Samsung': {
        'S22': [67000, 89000],
        'S23': [89000, 95000],
        'S24': [90000, 123000]
    },


    'OPPO': {
        'Reno': [23000, 25000],
        'X': [34000, 45000]
    },


    'One Plus': {
        '8': [23000, 34000],
        '9': [34000, 45000],
        '10': [45000, 52000]
    },


    'Google': {
        '8': [89000, 90000, 100000],
        '9': [123000, 145000, 163000]
    }
}
print(mobile)
```

Q1. What is the total number and names of the models in the brand Samsung?


```python
total=len(mobile["Samsung"])
print("Total:",total)
model =list(mobile["Samsung"].keys())
print("MOdel:",model)
```

Q2. What is the maximum price of Google 9?


```python
maximum = max(mobile["Google"]['9']) 
print("Max:",maximum)
```

Q3.What is the minimum price of Apple 14?


```python
minimum =min(mobile["Apple"]["14"])
print("Min:",minimum)
```

Q4. What is the maximum price of Samsung S24?

```python
maxi = max(mobile["Samsung"]["S24"])
print("Max:",maxi)
```

Q5. What is the average price of Apple 15?

```python
price = mobile["Apple"]["15"]
total = sum(price)
length =len(price)
average =total/length
print(average)
```

Q6. If you purchase 2 Samsung S24 base models and 3 Apple 14 highend mobiles, what will be the total price?


```python
sam = mobile["Samsung"]["S24"][0]
total_sam = sam*2
print("Total Price of 2 Samsung S24 base:",total_sam)
app = mobile["Apple"]["14"][2]
total_app =app*3
print("Total Price of 3 Apple 14 highend:",total_app)
print("Total price of 2 samsung s24 and 3 Apple 14:",total_sam+total_app)
```

Q7. If you purchase 20 Samsung S22 base models and 3 Google 9 highend mobiles, what will be the total price?


```python
sam_s22 = mobile["Samsung"]["S22"][0]
total_sam=sam_s22* 20
print("Total price of 20 SamSung S22 base:",total_sam)
goo_3=mobile["Google"]["9"][2]
total_goo =goo_3*3
print("Total price of 3 Google 9 highend:",total_goo)
print("Total price of 20 SamSung S22 base  and 3 Google 9 highend:",total_sam+total_goo)
```

Q8.How many brands are there in OPPO?


```python
op=mobile["OPPO"]
print("OPPO brands count:",len(op))
```

Q9. What are all the models available in OPPO?


```python
opp = mobile["OPPO"]
print("Models in OPPO:",list(opp.keys()))
```

Q10. How many price variants are there in Apple mobile?


```python
apple_mobiles = len(mobile["Apple"]["13"])+len(mobile["Apple"]["14"])+len(mobile["Apple"]["15"])
print("Total price variants in apple:",apple_mobiles)
```
