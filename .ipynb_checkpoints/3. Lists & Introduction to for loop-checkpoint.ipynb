
## 📦 What is a Python List?

Think of a **backpack** you carry around every day. Inside your backpack, you might have:

- 📚 A textbook
- 🍎 An apple
- 🔑 Keys
- 💳 Your wallet
- 📱 A smartphone
- 🗒️ A small note reminding you to "Buy milk"

Your backpack is special because:

- **It holds different kinds of items** *(heterogeneous)*.
- **Items are in a certain order**; you usually pack and unpack things sequentially *(ordered)*.
- You can **access items by position**—like grabbing the first thing you packed *(indexable)*.
- You can **add new items** anytime *(mutable)*.
- You can **remove items** you don't need anymore *(dynamic size)*.
- You can even have **multiple identical items** *(duplicates allowed)*.

| Feature    | `pop()`                      | `remove()`                      | `del`                         |
| ---------- | ---------------------------- | ------------------------------- | ----------------------------- |
| Removes by | Index (default last)         | Value (first occurrence)        | Index / slice / whole list    |
| Returns    | The removed element          | Nothing (`None`)                | Nothing                       |
| Errors     | `IndexError` if out of range | `ValueError` if value not found | `IndexError` if index invalid |
| Scope      | List method                  | List method                     | Python statement (general)    |


**Note: del keyword can permanently delete the instance of any datatype and we should exercise caution when using the del keyword**

### For preserving the list we need to use copy library which has methods like copy and deep copy which will preserve the original lists

````python
# Import the library
import copy
# Define list 1
list1 =[3,4,5,6,7,8]
print("list1:",list1)
# call library and function
list2 =copy.copy(list1)
print("list2",list2)
# Assign value to second list
list2[2] =9
# Print both
print("list1:",list1)
print("list2",list2)
````

### Shallow copy: copy()
Creates a new object, but the contents (nested objects) are still references to the original.

````python
# to create one more instance of a list
# Define list 1
odd_list=[1,3,[5,7],9]
# call library and function
mix =copy.copy(odd_list)
# Assign value to second list
mix[2][0]=6
# Print both
print(odd_list)
print(mix)
````

### deepcopy()
Creates a new object, and recursively copies all nested objects.

````python
# Define list 1
lk =[2,5,6,3,7,8]
# call library and function

ll =copy.deepcopy(lk)
# Assign value to second list
ll[1]=4
# Print both
print(lk)
print(ll)
````

````python

li =[2,3,4,[6,7],8]
# call library and function
l2 = copy.deepcopy(li)

# Assign value to second list
l2[3][0]=5
# Print both
print(li)
print(l2)
````

| Feature        | Shallow Copy       | Deep Copy                 |
| -------------- | ------------------ | ------------------------- |
| Top-level copy | ✅ Yes              | ✅ Yes                     |
| Nested objects | ❌ Still references | ✅ Fully copied            |
| Performance    | Faster             | Slower (recursive)        |
| Independence   | No (nested shared) | Yes (completely separate) |


📌 Rule of thumb:

* Use shallow copy if your data structure is flat (no nested mutable objects).

* Use deep copy if you have nested/multi-level structures (lists of lists, dicts of dicts, etc.).

### Simple For Loop Example

Let's loop through a list of fruits and print each one.

We are using:
- A list of fruits
- A `for` loop to iterate through each fruit
- A print statement to display a message

This is a beginner-friendly example to understand `for` loops in Python.


````python
# Define list of fruits
fruits = ['apple','mango','pineapple']

# Run loop
for fruit in ['apple','mango','pineapple']:
    print(fruit)

# Run loop with a message
for fruit in fruits:
    print('I like fruits such as:{}'.format(fruit))
````

````python
lst =[1,2,3,4]
for i in lst:
    print(i,end=" ")
````

````python
for i in lst:
    print('number order:{}'.format(i))
````

### Understanding `range()` in Python
The `range()` function generates a sequence of numbers.
- Used commonly in loops.
- Syntax: `range(start, stop, step)`
- If only one argument is given, it's treated as the stop.

```python



````python
for num in [4,5,6,7,8]:
    print(num, end = '\n')
for num in range(3,9):
    print(num , end = ' ')
````

````python
li =[1,2,3,4]
for i in li:
    print(i,end=" \n")
print("\n")    
for num in range(2,9):
    print(num)
````

Useful for controlled iteration, repeating tasks, or indexing.

### Understanding `enumerate()` in Python
`enumerate()` helps access both index and value during iteration.
- Syntax: `enumerate(iterable, start=0)`
- It returns pairs of (index, item)




````python
a,b = (3,5)
print(a,b)
````

````python
a,b = 3
print(a,b)
````

````python
# Define color list
colors = ["red", "green", "blue"]

# Run loop and print position and color
for position, color in enumerate(colors):
#     print(position,color)
    print('Position {} for {}'.format(position, color))
````

````python
l=["sai","harsha","harshini","teja"]
for i in enumerate(l):
    print(i)
    
````

````python
fruits = ['apple','mango','pineapple','grapes','banana']
for position,fruit in enumerate(fruits):
    print(f"fruits:{fruit} in position:{position}")
````

````python
ls =[1,2,3,4]
for p,i in enumerate(ls):
    print(p,i)
````

**Use Case:** Tracking index when looping over elements.
