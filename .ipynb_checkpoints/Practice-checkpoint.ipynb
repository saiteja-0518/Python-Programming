⃣   1 find Largest and Smallest in a List
Input: [3, 7, 2, 9, 4]
Output: Largest = 9, Smallest = 2



```python
n = int(input("Enter number of elements: "))
li = []

for i in range(n):
    val = int(input())
    li.append(val)

print("List :", li)
print("Largest :", max(li))
print("Smallest :", min(li))
```

2.Remove Duplicates from a List
Input: [1, 2, 2, 3, 4, 4, 5]
Output: [1, 2, 3, 4, 5]

```python
n = list(map(int,input('enter duplicates list values: ').split()))
un =[]
for i in n:
    if i not in un:
        un.append(i)
print("Original List:",n)
print("After Removing Duplicates:",un)
```

 3.Find Second Largest Number
Input: [10, 20, 4, 45, 99]
Output: 45


```python
n = list(map(int,input("enter list values").split()))
large = second =float('-inf')
for i in n:
    if i >large:
        second = large
        large =i
    elif i>second and i!=large:
        second =i
print("second largest:",second)
```

5.Count Even and Odd Numbers
Input: [1, 2, 3, 4, 5, 6]
Output: Even = 3, Odd = 3

```python
li = list(map(int,input("enter a numbers").split()))
even_count =0
odd_count =0
for i in li:
    if i %2==0:
        even_count+=1
    else:
        odd_count+=1
print(even_count)
print(odd_count)
```

9 Frequency of Elements in List
Input: [1, 2, 2, 3, 3, 3]
Output: {1:1, 2:2, 3:3}

```python
li = list(map(int,input("enter a list").split()))
freq ={}
for i in li:
    if i in freq:
        freq[i]+=1
    else:
        freq[i] =1
print(freq)
```

11 Merge Two Lists Without Duplicates
Input: [1, 2, 3], [2, 3, 4]
Output: [1, 2, 3, 4]

```python
l1 = list(map(int,input("enter a list").split()))
l2 = list(map(int,input("enter a list").split()))
merge =[]
for i in l1:
    if i not in merge:
        merge.append(i)
for i in l2:
    if i not in merge:
        merge.append(i)
print(merge)
```

⃣Inverted Right-Angled Triangle of Stars
Print the following pattern for n = 5:

```python
n = int(input("enter a num:"))
for i in range(n,0,-1):
    for j in range(i):
        print("*",end="")
    print()

```

```python
#Right-Angled Number Triangle
n = int(input())
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
#Inverted Number Triangle
n = int(input())
for i in range(n,0,-1):
    for j in range(1,i+1):
        print(j,end=" ")
    print()
```

```python
# Pyramid of Stars
n = int(input())
for i in range(1,n+1):
    for j in range(n-i):
        print(" ",end="")
    for k in range(2*i-1):
        print("*",end="")
    print()
    
```

```python
# Count Vowels and Consonants
s =input("Enter string: ")
v =0
c =0
for i in s:
    if i.isalpha():
        if i in "aeiouAEIOU":
            v+=1
        else:
            c+=1
print("Vowels",v)
print("Consonants",c)
```

```python
# prime number
li = list(map(int, input("Enter list elements: ").split()))
prime = []
for num in li:
    if num > 1:
        is_prime = True

        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            prime.append(num)

print(prime)
```
