Q1. Create a lambda function that converts temperature from Fahrenheit to Celsius.
 Input: temp = 96.8
 Output: 36

```python
celsius = lambda temp: round((temp-32)*5/9)
print(celsius(96.8))
```

Q2. Create a lambda function that finds the area of a circle.
 Input: radius = 5
 Output: 78.54

```python
import math
radius = lambda r : round(math.pi*(r**2),2)
print(radius(5))
```

Q3. Create a lambda function that converts American dollars to Indian rupees.
 Input: dollar_conversion = 45
 Output: 3950

```python
dollar = lambda r:(r*87)
print(dollar(45))
```

Q4. Create a lambda function that identifies child and adult (adult ≥ 18).
 Input: age = 10
 Output: Child

```python
adult = lambda age: "adult" if age>=18 else "child"
print(adult(10))
```

Q5. Create a lambda function that checks if a number is positive or negative.
 Input: -5
 Output: Negative

```python
num = lambda n:"positive" if n>0 else "Negative"
print(num(-5))
```

Q6. Write a lambda function that checks if a number is even or odd.
 Input: 7
 Output: Odd

```python
odd_even = lambda n: "even" if n%2==0 else "odd"
print(odd_even(7))
```

Q7. Create a lambda function that finds the maximum of two numbers.
 Input: 10, 25
 Output: 25

```python
maxi = lambda x,y: x if x>y else y
print(maxi(10,25))
```

Q8. Create a lambda function that checks if a person is eligible to vote (age ≥ 18).
 Input: 16
 Output: Not Eligible

```python
vote = lambda age : "Eligible" if age>=18 else "Not Eligible"
print(vote(16))
```

Q9. Write a lambda function that checks if a number is a multiple of 5.
 Input: 20
 Output: Yes


```python
multi = lambda x : "Yes" if x%5==0 else "No"
print(multi(20))
```

Q10. Write a lambda function to determine if a character is a vowel or consonant.
 Input: 'e'
 Output: Vowel

```python
character = lambda l : "Vowel" if l in "aeiou" else "constnant"
print(character('e'))
```

Q11. Create a lambda function that checks if a string is empty or not.
 Input: ""
 Output: Empty

```python
m = lambda s: "Empty" if s=="" else "Not empty"
print(m(""))
```

Q12. Write a lambda function that checks if a number lies between 10 and 50 (inclusive).
 Input: 35
 Output: Yes

```python
j = lambda x: "Yes" if x >=10 and x<=50 else "No"
print(j(35))
```

Q13. Create a lambda function that returns Pass if marks ≥ 40, otherwise Fail.
 Input: 28
 Output: Fail

```python
h = lambda s: "pass" if s>=40 else "Fail"
print(h(28))
```

Q14. Write a lambda function that checks if two numbers are equal or not.
 Input: 15, 20

```python
r = lambda x,y :"Equal" if x==y else "Not equal"
print(r(15,20))
```

Q15. Define a lambda function that converts odd numbers to even.
 Input: 5
 Output: 6

```python
k = lambda s: s if s%2==0 else s+1
print(k(5))
```
