## Conditional Statements


1) What's Your Grade?
Question: You got a score from 0 to 100. Figure out your letter grade (A, B, C, D, F).
A: 90-100
B: 80-89
C: 70-79
D: 60-69
F: Below 60
Input: A whole number (your score).
Output: Your grade (e.g., A, B, F).
Sample Input:
92
Sample Output:
A






```python
marks =int(input("enter the marks:"))
if marks>=90 and marks<=100:
    print("Your Grade A")
elif marks >=80 and marks <=89:
    print (" Your Grade B")
elif marks >=70 and marks <=79:
    print(" Your Grade C")
elif marks >=60 and marks <=69:
    print("Your Grade D")
elif marks<60:
    print("Your Grade F")    
    
```

2) What Kind of Triangle?
Question: You have three stick lengths. Can you make a triangle?
If not, say "Invalid". (Remember: two sides must be longer than the third side combined).
If yes, what kind is it? "Equilateral" (all sides same), "Isosceles" (two sides same), or "Scalene" (all sides different).
Input: Three whole numbers (side lengths).
Output: The triangle type (e.g., Scalene Right).
Sample Input:
3 4 5
Sample Output:
Scalene


```python
a = int(input("enter the 1st side"))
b= int(input("enter the 2nd side"))
c =int(input("enter the 3rd side"))
if a+b>c and b+c>a and c+a>b:
    if a == b and b==c and c==a:
        print("The triangle type:Equilateral")
    elif a ==b or c==a or b==c:
        print("The triangle type: Isosceles")
    elif a !=b and b != c and c!=a:
        print("The triangle type Scalene")
else:
    print("Invalid")
```

3) Where’s the Dot?
Question: You are given a point (x, y). Decide where it lies:
Origin: x = 0 and y = 0.
Example: (0, 0) → Origin
X-Axis: y = 0 but x ≠ 0.
Example: (5, 0) → X-Axis; (-7, 0) → X-Axis
Y-Axis: x = 0 but y ≠ 0.
Example: (0, 3) → Y-Axis; (0, -9) → Y-Axis
Q1: x > 0 and y > 0 (top-right).
Example: (4, 2) → Q1
Q2: x < 0 and y > 0 (top-left).
Example: (-3, 6) → Q2
Q3: x < 0 and y < 0 (bottom-left).
Example: (-2, -5) → Q3
Q4: x > 0 and y < 0 (bottom-right).
Example: (7, -1) → Q4
Input: Two integers x y.
Output: One of: Origin, X-Axis, Y-Axis, Q1, Q2, Q3, Q4.
Sample Input:
-5 0
Sample Output:
X-Axis




```python
x = int(input("enter integer value1:"))
y =  int(input("enter integer value2:"))
if x ==0 and y ==0:
    print("Origin")
elif y ==0 and x != 0:
    print("X-Axis")
elif x==0 and y !=0:
    print("Y-Axis")
elif x>0 and y>0:
    print("Q1")
elif x<0 and y >0:
    print("Q2")
elif x<0 and y<0:
    print("Q3")
elif x>0 and y<0:
    print("Q4")
```

4) Your Power Bill
Question: Calculate your electricity bill based on how many units you used.
First 100 units: $1.50 per unit.
Next 200 units (from 101 to 300): $2.50 per unit.
Any units over 300: $4.00 per unit.
Add a fixed $50 meter charge.
Input: A whole number (units consumed).
Output: Your total bill amount.
Sample Input:
350
Sample Output:
900.00


```python
units = int(input("enter the units:"))
if units<=100:
    print(units*1.50+50)
elif units <=300:
    print((100*1.50)+((units -100)*2.50)+50)
else:
    print((100*1.50)+(200*2.50)+((units-300)*4.00)+50)5
```
