# Tuple
- Collection which can be indexed but immutable, (apple, 250)
- So once we declare the tuple, we can not edit the data
- tuple will be declared using `()`

### tuple declaration

```python
# create a tuple
num_tuple = (2,3,5,6,7)
print(num_tuple)

#type of variable
print(type(num_tuple))

# length of tuple
print(len(num_tuple))
```

```python
# create a tuple with single item
int_tuple =(6)
print(int_tuple)
#type of variable
print(type(int_tuple))

# length of tuple
print(len(int_tuple))
```

```python
# create a tuple with single item
int_tuple =(6,)
print(int_tuple)
#type of variable
print(type(int_tuple))

# length of tuple
print(len(int_tuple))
```

### How to add elements in tuple?

```python
print(int_tuple)
# create another tuple
num_tuple =(7,8,9)

# add the tuple
T1 =int_tuple + num_tuple
print(T1)
T2 =num_tuple + int_tuple
print(T2)
```

## Built in Methods

* count() -> number of occurence of a character
* index() -> position of the variable

```python
# count()
t = (3,4,5,2,4,5,6,3,9,3)
print(f"tup:{t}")

# count Occurence of 3
print(f"Occurence of 3 :{t.count(3)}")
```

```python
# Find the index value of 7 
r =(2,3,4,1,7,8,7,0,7)
print(f"tup:{r}")
# index of 7
i =r.index(7)
print(f"index value of 7 :{i}")
# repeted value 
re = r.index(7,i+1)
print(f"repated vlaue :{re}")
```

## sets

- **Sets are like lists in that you use them for storing a collection of elements. Unlike lists, however, the elements in a set are nonduplicates and are not placed in any particular order.**
- **with { } we declare a set**
- **sets are unordered and unindexed**
- **set will not allow duplicate elements**

```python
# sets declaration
set1 = {1.8,2.5,3.4,}
print(set1)
print(type(set1))

# sets decalrtion with differnet datatyes
sq = {2.3,5.6,7,"name",True}
print(sq)
print(type(sq))
```

```python
# Creating empty set
empty={}
print(empty)
print(type(empty))
```

***NOTE:*** empty curly leads to empty dictionary rather than empty set

```python
empty_set=set()
print(empty_set)
print(type(empty_set))
```

```python
# creating set using list
```

```python
l1 =[1,2,2,3,3,4,5,5]
print(l1)
# type converstion
set1 =set(l1)
print(set1)
```

```python
# remove duplicate 
li = [1,2,3,4,5,4,5]
set1 =set(li)
print(list(set1))
```

```python
# nested list
l1 =[2,3,4,[3,4,5]]
print(l1)
print(set(l1))
```

# Other set operations
- 1. union()
- 2. intersection()
- 3. differeence()

### union()

```python
#Sets union, intersection and difference
```

```python
#create a setA and setB
setA = {1,2,3,4,5}
setB ={4,5,6,7,8}

#union
print(f"setA union setB: {setA | setB}")
print(f"setB union setA: {setB | setA}")
print(f"union with methods:{setA.union(setB)}")
print(f"union with methods:{setB.union(setA)}")
```

### intersection()

```python
#Sets union, intersection and difference
A ={1,2,3,4,5}
B ={4,5,6,7,8}
print(f"A intersection B:{A&B}")
print(f"A intersection B:{B&A}")
print(f"A intersection B:{A.intersection(B)}")
print(f"A intersection B:{B.intersection(A)}")
```

### difference()

### symmetric difference

```python
#Previous class notes
x={1,2,3,4,5}
y ={4,5,6,7,8} 
#Symmetric difference prints values of A and B and eliminates the common
#elements from both the sets
print(x^y)
print(y^x)
#This can also be written as below notation
print((x|y)-(x&y))
print((y|x)-(y&x))
#Both these gives same output
print(x.symmetric_difference(y))
print(y.symmetric_difference(x))
```

# 2. Manipulating and Accessing Sets
- You can add an element to a set or remove an element by using the add(e) or remove(e) method. You can use the len, min, max, and sum functions on a set, and a for loop to traverse all elements in a set. You can use the in or not in operator to determine whether an element is in the set.

```python
 # Use function len
 # Use max
 # Use min
 # Use sum
s1 ={1,2,3,4,5}
# length
print("length is:",len(s1))
# maximum
print("max value is:",max(s1))
# minimum
print("min value is:",min(s1))
# sum of values
print("sum of s1:",sum(s1))
```
