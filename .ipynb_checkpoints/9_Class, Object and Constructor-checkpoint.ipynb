# What is a Class and Object?

## 1. Class: The Blueprint

A **class** is a blueprint or template used to create objects.  
It defines the structure (what data the object will hold) and behavior (what actions it can perform), but it does not hold any actual data by itself.

Think of a class as a **recipe**.  
The recipe defines ingredients and steps, but the actual cake isn’t made until you follow the recipe.


```python
# Define a class

class ClassName:
    
    def __int__(self):
        # attributes
        pass
    
    def method_name(self):
        # functions/actions/ behaviour
        pass
    
```

```python
### Syntax of a Class in Python:

class ClassName:
    def __init__(self):
        # attributes
        pass

    def method_name(self):
        # behavior
        pass
```

## 2. Object: The Instance

An **object** is a real-world instance of a class — something that actually exists in memory and can store data and perform operations.

Using our earlier bakery example:

- The `Customer` class is the blueprint.
- `customer1 = Customer("Deva", "9876543210")` is an **object** — an actual customer with real details.

You can create **multiple objects** from a single class, each holding its own data.


## 3. Example: Defining and Using a Class

Let’s define a `Customer` class with attributes and methods.


```python
# Define Customer Class
class Customer:
    # attributes/ characteristics
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    def greet(self):
        print(f"Welcome {self.name}")
    
```

```python
# call class/ create instance(object)
Customer()
```

```python
Customer('ram', 7534898234)
```

```python
# Create object or Instatiation
customer1 = Customer('ram', 7534898234)
```

```python
print(customer1.greet())
print(customer1.name)
```

```python
class Customer:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def greet(self):
        print(f"Welcome, {self.name}!")

# Creating objects
customer1 = Customer("Deva", "9876543210")
customer2 = Customer("Yash", "9988776655")

# Using object methods
customer1.greet()
customer2.greet()
```

```python
customer1.greet()
customer2.greet()
```

### Explaination:
- Customer is the class
- customer1 and customer2 are objects of that class
- name and phone are instance variables
- greet() is a method (behavior)

# The `__init__` Method and Instance Variables

## 1. What is the `__init__` Method?

In Python, the `__init__` method is a **special method** that is automatically called when a new object is created from a class.

Its primary role is to **initialize the object** — in other words, set up the object with initial data (also called attributes or instance variables).

You can think of it as a **constructor** in other programming languages.


```python
### Syntax:
class ClassName:
    def __init__(self, param1, param2):
        self.param1 = param1
        self.param2 = param2

```

- self refers to the current object being created


- The parameters passed to __init__ are assigned to the object’s attributes


## 2. What are Instance Variables?

Instance variables are variables that belong to a specific **object** (or **instance**) of a class.  
They are created inside the `__init__` method using `self`.  
Each object gets its own copy of instance variables.

### Create an Employee class that shows the output as name, department and salary of an employee.

```python
class Employee:
    def __init__(self,name, department, salary):
        self.name = name
        self.department = department
        self.salary = salary
    
    def show_info(self):
        print(f"Name of Employee: {self.name}, Department: {self.department}, Salary: {self.salary}")
        
```

```python
employee1 = Employee('deva', 'HR', 120000)
```

```python
employee1.show_info()
```

```python
employee1.name
```

```python
class Employee:
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.salary = salary

    def show_info(self):
        print(f"Employee: {self.name}, Department: {self.department}, Salary: ₹{self.salary}")

# Creating objects
emp1 = Employee("Aarav", "HR", 35000)
emp2 = Employee("Neha", "IT", 55000)

emp1.show_info()
emp2.show_info()
```

### Explanation:

- The `__init__` method **assigns the values** when the object is created.
- Each object (`emp1`, `emp2`) holds its own data independently. 
  - For example, `emp1` will have its own `name` and `phone`, and `emp2` will have different values for these attributes.



| Element       | Meaning                                                     |
|---------------|-------------------------------------------------------------|
| `__init__()`  | Constructor method called automatically at object creation |
| `self.variable` | Instance variable unique to the object                    |
| `self`        | Refers to the current instance of the class                 |


# Understanding Methods: Instance Methods, Class Methods, and Static Methods

In Python classes, **methods** are functions defined inside a class. They define the behavior of objects and help you interact with object data.  
Python provides three types of methods inside a class:

- **Instance Methods**
- **Class Methods**
- **Static Methods**

Each type has a different purpose and behaves differently based on how it accesses data.


## 1. Instance Methods

These are the most commonly used methods. They operate on individual **object data** (instance variables) and take `self` as the first parameter.

```python
### Example

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display_info(self):  # Instance method
        print(f"Name: {self.name}, Marks: {self.marks}")

# Usage
s1 = Student("Riya", 87)
s1.display_info()
```

### Explaination

- self refers to the object calling the method
- Instance methods can access and modify the object’s data


## 2. Class Methods

Class methods operate on the **class itself** rather than individual objects.  
They take `cls` as the first parameter and are declared using the `@classmethod` decorator.

Use class methods when you need to:

- Work with **class-level data**
- Provide **alternate constructors** for creating objects


```python
### Example:

class Product:
    company = "Innomatics Research Labs"  # Class variable

    def __init__(self, name, price):
        self.name = name
        self.price = price

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name

# Usage
Product.change_company("IRL Pvt Ltd")
print(Product.company)
```

```python
class Employee:
    company = 'TCS'
    
    def __init__(self, name, department, salary):
        self.name = name
        self.department = department
        self.salary = salary

    def show_info(self):
        print(f"Employee: {self.name}, Department: {self.department}, Salary: ₹{self.salary}")

    
    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name
```

```python
Employee.company
```

```python
Employee.change_company('Wipro')
```

```python
Employee.company
```

```python
# Creating objects
emp1 = Employee("Aarav", "HR", 35000)
emp2 = Employee("Neha", "IT", 55000)

emp1.show_info()
emp2.show_info()
```

### Explanation

- cls refers to the class itself
- Useful for modifying class variables shared across all objects


## 3. Static Methods

Static methods do not take `self` or `cls` as the first argument.  
They behave like **regular functions**, but they belong to the class namespace and are logically grouped with the class.

Use static methods for:

- **Utility/helper functions**
- Operations related to the class but not dependent on **instance** or **class data**

They are declared using the `@staticmethod` decorator.


```python
## Example:

class Calculator:
    @staticmethod
    def add(x, y):
        print(x + y)

    @staticmethod
    def greet():
        print("Welcome to the Calculator App")
```

```python
# Usage
Calculator.add(2,3)
Calculator.greet()
```

# Method Type Comparison

| Method Type     | Decorator      | First Argument | Can Access Instance Data | Can Access Class Data | Use Case                                |
|-----------------|----------------|----------------|---------------------------|------------------------|-----------------------------------------|
| **Instance Method** | No decorator   | `self`         | Yes                       | Yes (via `self`)        | Working with object-specific data      |
| **Class Method**    | `@classmethod` | `cls`          | No                        | Yes                     | Working with class variables or settings|
| **Static Method**   | `@staticmethod`| None           | No                        | No                      | Utility/helper methods                 |


# Examples for the above Topics

## Use Case - 1 : Products Information

### Understanding the Use Case
- In a bakery, there are different products like cakes, pastries, and breads.
- Each product has two important pieces of information:
   - Name (like "Chocolate Cake")
   - Price (like ₹500)
- Instead of managing names and prices separately, we can bundle them together inside a Product class.
- Each item you sell will be an object of this class.

```python
class Product:
    def __init__(self, name, price):
        self.name = name  # Atrributes / Instance variables
        self.price = price  # Attributes/ Instance variables
    
    def product_info(self):
        return f"Product Name: {self.name}, Price: ₹{self.price}"
```

```python
product1 = Product('cake', 250)
product2 = Product('Chocolate', 150)

print(product1.product_info())
print(product2.product_info())
```

```python
# Defining a Product class
class Product:
    def __init__(self, name, price):
        self.name = name       # Attribute: name of the product
        self.price = price     # Attribute: price of the product

    def display_info(self):
        print(f"Product: {self.name}, Price: ₹{self.price}")

# Creating Product objects
cake = Product("Chocolate Cake", 500)
bread = Product("Garlic Bread", 150)

# Using object methods
cake.display_info()
bread.display_info()
```

## Use Case - 2 : Orders

### Understanding the Use Case
- In a bakery, customers don't just buy one item — they usually order multiple quantities of a product.
- **For example:**
    - 4 pastries
    - 6 donuts
- Each Order needs to store:
    - Product Name (like "Pastry")
    - Quantity (how many pieces)
    - Unit Price (price for one piece)

- Instead of keeping track of this manually, we can create an Order class.
- Each customer's order will be an object of this class.

```python
# Defining a Product class
class Product:
    def __init__(self, name, unit_price, quantity):
        self.name = name       # Attribute: name of the product
        self.unit_price = unit_price     # Attribute: price of the product
        self.quantity = quantity
        
    def display_info(self):
        print(f"Product: {self.name}, Price: ₹{self.unit_price * self.quantity}")

# Creating Product objects
cake = Product("Chocolate Cake", 500, 2)
bread = Product("Garlic Bread", 150, 5)

# Using object methods
cake.display_info()
bread.display_info()
```

```python
# Defining an Order class
class Order:
    def __init__(self, product_name, quantity, unit_price):
        self.product_name = product_name    # Attribute: name of the product ordered
        self.quantity = quantity            # Attribute: how many units ordered
        self.unit_price = unit_price         # Attribute: price per unit

    def total_price(self):
        return self.quantity * self.unit_price  # Behavior: calculating total price

# Creating Order objects
order1 = Order("Pastry", 4, 80)
order2 = Order("Donut", 6, 50)

# Calculating and displaying total price
print(f"Total for {order1.product_name}: ₹{order1.total_price()}")
print(f"Total for {order2.product_name}: ₹{order2.total_price()}")
```
