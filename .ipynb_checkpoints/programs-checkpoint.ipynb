11.Write a Python program to check whether a given number is prime or not.

```python
num = int(input())
if num <=1:
    print(f"{num}is not a prime number")
else:
    is_prime =True
    for i in range(2,int(num**0.5)+1):
        if num %i ==0:
            is_prime =False
            break
    if is_prime:
        print(f"{num} is a prime number")
    else:
        print(f"{num} is not a prime number ")
```

12.Write a Python program to count the number of digits in a given integer.

13.Write a Python program to check whether a given number is a palindrome.

```python
num =int(input())
original =num
reverse =0
while num >0:
    last =num %10
    reverse =(reverse *10) +last
    num //=10
    
if original ==reverse:
    print(f"{original} is a palindrome")
else:
     print(f"{original} is not a palindrome")
```

14 Student Grade Calculator

```python
name =input("Enter Name:")
mark1 =int(input("Enter Mark1"))
mark2 =int(input("Enter Mark2"))
mark3 =int(input("Enter Mark3"))
mark4 =int(input("Enter Mark4"))
mark5 =int(input("Enter Mark5"))
total = mark1+mark2+mark3+mark4+mark5
avg =total/5
if avg >=90:
    grade ="A+"
elif avg>=80:
    grade ="A"
elif avg>=70:
    grade ="B"
elif avg>=60:
    grade ="C"
else:
    grade ="Fail"
print(f"Student Name:{name}")
print(f"Total: {total}")
print(f"Average: {avg}")
print(f"Grade: {grade}")
```

15 .Remove Duplicate Elements from a List
Write a Python program to:
Accept N elements from the user.
Store them in a list.
Remove duplicate elements without using set().
Display the original list and the new list.

```python
n =int(input("Enter number of elements: "))
original =[]
for i in range(n):
    val =int(input())
    original.append(val)
unique=[]
for i in original:
    if i not in unique:
        unique.append(i)
print("Original List:")
print(original)
print("\nUnique List:")
print(unique)
```

16. Employee Management Using OOP

```python
class Employee:
    def __init__(self,employee_id,employee_name,department,salary):
        self.employee_id=employee_id
        self.employee_name=employee_name
        self.department =department
        self.salary =salary   
        
    def display(self):
        print("Employee Details")
        print(f"Employee ID :{self.employee_id}",)
        print(f"Employee Name :{self.employee_name}",)
        print(f"Department :{self.department}",)
        print(f"Monthly Salary :{self.salary}",)
        
    def annual_salary(self):
        print(f" Annual Salary: {self.salary*12}")
        
employee_id = int(input("Employee ID :"))    
employee_name = input("Employee Name :")
department = input("Department :")    
salary = int(input("Monthly Salary :"))  

emp = Employee(employee_id,employee_name,department,salary)
emp.display()
emp.annual_salary()
```

17 Write a Python program to check whether a number is an Armstrong number.

```python
num = int(input("enter a number:"))
digit = str(num)
l =len(digit)
total = sum(int(d)**l for d in digit)
if total ==num:
    print(f"{num} armstrong")
else:
    print(f"{num}is not armstrong")
    
```

19. GCD (HCF) of Two Numbers

```python
n1 = int(input("enter num1"))
n2 = int(input("enter num2"))
a = n1
b =n2
while b!=0:
    a,b =b,a%b
gcd =a
print(gcd)
```

20 Write a Python program to find the LCM of two numbers.

```python
n1 =int(input("enter num1"))
n2= int(input("enter num2"))
a =n1
b =n2
while b!=0:
    a,b = b,a%b
gcd =a
lcm =(n1* n2)//gcd
print(lcm)
```

21.Student Result Analysis

Write a Python program to:

Accept the student's name
Accept the marks of 5 subjects
Calculate:
Total
Average
Highest mark
Lowest mark
Display the result:
Pass if every subject mark is 35 or above
Otherwise display Fail
Display the grade based on the average:

```python
name = input("enter a name:")
s1 = int(input("enter a subject1 marks:"))
s2 = int(input("enter a subject2 marks:"))
s3 = int(input("enter a subject3 marks:"))
s4 = int(input("enter a subject4 marks:"))
s5 = int(input("enter a subject5 marks:"))
total = s1+s2+s3+s4+s5
avg =total/5
highest = max(s1,s2,s3,s4,s5)
lowest = min(s1,s2,s3,s4,s5)
if s1>35 and s2>35 and s3 >35 and s4>35 and s5>35:
    result="Pass"
else:
    result="Fail"
if avg >=90:
    grade ="A+"
elif avg >=80:
    grade ="A"
elif avg >=70:
    grade ="B"
elif avg >=60:
    grade ="C"
else:
    grade ="D"
print(f"Student name:{name}")
print(f"Total:{total}")
print(f"Average : {avg}")
print(f"Highest : {highest}")
print(f"Lowest : {lowest}")
print(f"Result : {result}")
print(f"Grade : {grade}")
```

22.Word Frequency Counter

Write a Python program that:

Accepts a sentence from the user.
Converts the sentence to lowercase.
Splits it into words.
Stores the frequency of each word in a dictionary.
Displays each word with its count.

```python
s =input("enter a sentence")
word =s.lower().split()
count ={}
for i in word:
    if i in count:
        count[i] +=1
    else:
        count[i]=1
for i, counts in count.items():
    print(f"{i} : {counts}")


        
```
