```python
from abc import ABC, abstractmethod


class Account(ABC):

    def __init__(self, account_number, customer_name, balance):
        self.account_number = account_number
        self.customer_name = customer_name
        self.__balance = balance

    # -----------------------------
    # Getter Method
    # -----------------------------
    def get_balance(self):
        return self.__balance

    # -----------------------------
    # Deposit
    # -----------------------------
    def deposit(self, amount):

        if amount <= 0:
            print("Invalid Deposit Amount")
            return

        self.__balance += amount
        print("Amount Deposited Successfully")
        print("Current Balance :", self.__balance)

    # -----------------------------
    # Withdraw
    # -----------------------------
    def withdraw(self, amount):

        if amount <= 0:
            print("Invalid Withdrawal Amount")
            return

        if amount > self.__balance:
            print("Insufficient Balance")
            return

        self.__balance -= amount

        print("Amount Withdrawn Successfully")
        print("Current Balance :", self.__balance)

    # -----------------------------
    # Display Details
    # -----------------------------
    def display_details(self):

        print("-" * 40)
        print("Account Number :", self.account_number)
        print("Customer Name  :", self.customer_name)
        print("Balance        :", self.get_balance())

    # -----------------------------
    # Abstract Method
    # -----------------------------
    @abstractmethod
    def calculate_interest(self):
        pass


# ==================================================
# Savings Account
# ==================================================

class SavingsAccount(Account):

    def __init__(self,
                 account_number,
                 customer_name,
                 balance,
                 interest_rate):

        super().__init__(
            account_number,
            customer_name,
            balance
        )

        self.interest_rate = interest_rate

    def calculate_interest(self):

        interest = (
            self.get_balance() *
            self.interest_rate
        ) / 100

        print("Interest Earned :", interest)

    def display_details(self):

        super().display_details()

        print("Account Type   : Savings")
        print("Interest Rate  :", self.interest_rate, "%")

        print("-" * 40)


# ==================================================
# Current Account
# ==================================================

class CurrentAccount(Account):

    def __init__(self,
                 account_number,
                 customer_name,
                 balance,
                 overdraft_limit):

        super().__init__(
            account_number,
            customer_name,
            balance
        )

        self.overdraft_limit = overdraft_limit

    def calculate_interest(self):

        print("Interest is not applicable for Current Accounts.")

    def display_details(self):

        super().display_details()

        print("Account Type     : Current")
        print("Overdraft Limit  :", self.overdraft_limit)

        print("-" * 40)

accounts = []

def search_account(account_number):

    for account in accounts:

        if account.account_number == account_number:
            return account

    return None

while True:

    print("\n========== BATCH 542 SMART BANK ==========")
    print("1. Create Savings Account")
    print("2. Create Current Account")
    print("3. Deposit Money")
    print("4. Withdraw Money")
    print("5. View Account Details")
    print("6. Calculate Interest")
    print("7. Display All Accounts")
    print("8. Exit")
    print("================================")

    choice = int(input("Enter your choice: "))

    # ----------------------------------
    # Create Savings Account
    # ----------------------------------

    if choice == 1:

        account_number = input("Enter Account Number: ")
        customer_name = input("Enter Customer Name: ")
        balance = float(input("Enter Opening Balance: "))
        interest_rate = float(input("Enter Interest Rate: "))

        account = SavingsAccount(
            account_number,
            customer_name,
            balance,
            interest_rate
        )

        accounts.append(account)

        print("Savings Account Created Successfully.")

    # ----------------------------------
    # Create Current Account
    # ----------------------------------

    elif choice == 2:

        account_number = input("Enter Account Number: ")
        customer_name = input("Enter Customer Name: ")
        balance = float(input("Enter Opening Balance: "))
        overdraft_limit = float(input("Enter Overdraft Limit: "))

        account = CurrentAccount(
            account_number,
            customer_name,
            balance,
            overdraft_limit
        )

        accounts.append(account)

        print("Current Account Created Successfully.")

    # ----------------------------------
    # Deposit
    # ----------------------------------

    elif choice == 3:

        account_number = input("Enter Account Number: ")

        account = search_account(account_number)

        if account:

            amount = float(input("Enter Deposit Amount: "))

            account.deposit(amount)

        else:

            print("Account Not Found.")

    # ----------------------------------
    # Withdraw
    # ----------------------------------

    elif choice == 4:

        account_number = input("Enter Account Number: ")

        account = search_account(account_number)

        if account:

            amount = float(input("Enter Withdrawal Amount: "))

            account.withdraw(amount)

        else:

            print("Account Not Found.")
    # ----------------------------------
    # View Account Details
    # ----------------------------------

    elif choice == 5:

        account_number = input("Enter Account Number: ")

        account = search_account(account_number)

        if account:

            account.display_details()

        else:

            print("Account Not Found.")
    # ----------------------------------
    # Calculate Interest
    # ----------------------------------

    elif choice == 6:

        account_number = input("Enter Account Number: ")

        account = search_account(account_number)

        if account:

            account.calculate_interest()

        else:

            print("Account Not Found.")
    # ----------------------------------
    # Display All Accounts
    # ----------------------------------

    elif choice == 7:

        if len(accounts) == 0:

            print("No Accounts Available.")

        else:

            print("\n========== ALL ACCOUNTS ==========\n")

            for account in accounts:

                account.display_details()
    # ----------------------------------
    # Exit
    # ----------------------------------

    elif choice == 8:

        print("\nThank you for using Smart Banking System.")

        break
    
```
