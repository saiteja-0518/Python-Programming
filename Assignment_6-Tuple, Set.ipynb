Q19. Given sentence = "apple banana apple cherry banana", split it into a list, convert it to a set to find unique words, and use a for loop to print each unique word.


```python
sentence = "apple banana apple cherry banana"
w = sentence.split()
t = set(w)
for w in t:
    print(w)
```
