```python
#Q10. Create a nested list [1, 2, [3, 4]]. Create a shallow copy of it. Add the number 5 to the inner list of the copy and print both the original and the copied list.
```

```python
import copy
lst =[1, 2, [3, 4]]
lst2=copy.copy(lst)
lst2[2].append(5)
print(lst)
print(lst2)
```

```python
# Q11. Create a nested list [1, 2, [3, 4]]. Create a deep copy of it. Add the number 5 to the inner list of the copy and print both. How does this result differ from the shallow copy in Q10?
```

```python
import copy
i =[1, 2, [3, 4]]
k =copy.deepcopy(i)
k[2].append(5)
print(i)
print(k)
#deep copy creates a seperate copy memory
#shallow copy shares inner 
```

```python
# Q16. Create a list ["a", "b", "c", "d"]. Reverse the list in-place and then use enumerate to print each element with a position number starting from 1.
```
