Q1. Write a function greet(name, msg="Welcome!") that prints a greeting. Call it with and without the msg argument.
 Input: greet("Alice")
 Output: Welcome! Alice
 Input: greet("Bob", "Hello")
 Output: Hello Bob


```python
def greet(name,msg="Welcome"):
    print(msg,name)
greet("Alice")
greet("Bob","Hello")
```

Q2. Write a function power(num, exp=2) that returns num raised to the power exp. Call it with one and two arguments.
 Input: power(3)
 Output: 9
 Input: power(2, 4)
 Output: 16


```python
def power(num,exp=2):
    return num ** exp
print(power(3))
print(power(2,4))
```

Q3. Write a function multiply(*numbers) that returns the product of all numbers passed to it. Call it with 3 and 5 arguments.
 Input: multiply(2, 3, 4)
 Output: 24

```python
def multiply(*num):
    r =1
    for i in num:
        r*=i
    return r
print(multiply(2,3,4))  
print(multiply(1,2,3,4,5))
```

Q4. Write a function concat_strings(*args) that joins all string arguments with a space and returns the result. Call it with 2 and 4 strings.
 Input: concat_strings("Hello", "World")
 Output: "Hello World"
 Input: concat_strings("Python", "is", "fun", "today")
 Output: "Python is fun today"

```python
def concat_strings(*args):
    return " ".join(args)
print(concat_strings("Hello", "World"))
print(concat_strings("Python", "is", "fun", "today"))
```

Q6. Write a function max_of_numbers(*nums) that returns the maximum value among numbers passed. Call it with 3 and 6 numbers.
 Input: max_of_numbers(3, 7, 1)
 Output: 7
 Input: max_of_numbers(5, 2, 9, 4, 8, 1)
 Output: 9

```python
def max_of_numbers(*nums):
    m =nums[0]
    for i in nums:
        if i >m:
            m=i
    return m
print(max_of_numbers(3, 7, 1))
print(max_of_numbers(5, 2, 9, 4, 8, 1))
```

8. Write a function print_kwargs(**kwargs) that prints all key-value pairs passed as keyword arguments. Call it with 1 and 3 key-value pairs.
 Input: print_kwargs(name="Alice")
 Output: name: Alice
 Input: print_kwargs(name="Bob", age=25, city="Delhi")
 Output:
name: Bob
age: 25
city: Delhi


```python
def print_kwargs(**kwargs):
    for k in kwargs:
        print(f"{k}: {kwargs[k]}")
print_kwargs(name = "Alice")
print_kwargs(name="Bob", age=25, city="Delhi")
```

Q9. Write a function greet_friends(*friends, msg="Hi") that prints the greeting message followed by each friend's name. Call it with 2 and 4 friends, with and without msg.
 Input: greet_friends("Alice", "Bob")
 Output:
Hi Alice
Hi Bob

Input: greet_friends("Tom", "Jerry", "Spike", "Tyke", msg="Hello")
 Output:
Hello Tom
Hello Jerry
Hello Spike
Hello Tyke

Q10. Write a function sum_numbers(*numbers, start=0) that returns the sum of all numbers plus start. Call it with numbers only and with start value.
 Input: sum_numbers(1, 2, 3)
 Output: 6
 Input: sum_numbers(1, 2, 3, start=5)
 Output: 11

```python
def sum_numbers(*numbers, start=0):
    add = start
    for i in numbers:
        add +=i
    return add
print(sum_numbers(1, 2, 3))
print(sum_numbers(1, 2, 3, start=5))
```

Q12. Write a function list_to_string(*items, sep="-") that joins all items using the separator. Call it with default and custom separator.
 Input: list_to_string("a", "b", "c")
 Output: "a-b-c"
 Input: list_to_string("1", "2", "3", sep=":")
 Output: "1:2:3"


```python
def list_to_string(*items, sep="-"):
    return sep.join(items)
    
print(list_to_string("a", "b", "c"))
print(list_to_string("1", "2", "3", sep=":"))   
```
