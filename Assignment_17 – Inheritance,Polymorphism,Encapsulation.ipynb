# Part B: Polymorphism

Q16. Write a Python program that demonstrates polymorphism using a function area() that works for both circle and rectangle.

```python
import math
class Cricle:
    def __init__ (self,radius):
        self.radius =radius
    def area(self):
        return math.pi*self.radius**2
class Rectangle:
    def __init__(self,height,width):
        self.height = height
        self.width =width
    def area(self):
        return self.height * self.width
c = Cricle(5)
r = Rectangle(3,8)
print(c.area())
print(r.area())
```

### Part C: Encapsulation

Q21. Create a class BankAccount with private variables _balance. Provide methods to deposit and withdraw money, and show how encapsulation controls access.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0      
    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited {amount}, Balance: {self.__balance}")
    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance")
        else:
            self.__balance -= amount
            print(f"Withdrawn {amount}, Balance: {self.__balance}")
    def get_balance(self):
        print(f"Balance: {self.__balance}")
b = BankAccount()
b.deposit(1000)
b.withdraw(500)
b.withdraw(700)
b.get_balance()
```

Q22. Create a class Laptop with private attributes _brand and _price. Use encapsulation to print details only if the user has proper access.


```python
class Laptop:
    def __init__(self, brand, price):
        self.__brand = brand
        self.__price = price
    def get_details(self, access):
        if access == "admin":
            print(f"Brand: {self.__brand}, Price: {self.__price}")
        else:
            print("Access Denied")
l = Laptop("Dell", 50000)
l.get_details("admin")
l.get_details("user")
```
