Q1. Search for a character ch in a string s. If found, return "Found". If not found, return "Not found" (use for-else).
Input: s = "hello", ch = "e" Output: "Found"


```python
s = "hello"
ch ="e"
for i in s:
    if i ==ch:
        print("Found")
        break
else:
    print("Not Found")
```

Q2. Given a list of student names, check if "Alice" exists in the list. If found, return "Alice present". Otherwise, return "Alice absent" (use for-else).
 Input: ["John", "Bob", "Clara"]
 Output: "Alice absent"



```python
p = ["John", "Bob", "Clara"]
for i  in p:
    if i =="Alice":
        print("Alice Present")
        break
else:
    print("Alice absent")
```

Q3. Count how many times a number appears in a nested list.
 Input: [[1, 2, 3], [4, 2, 6], [7, 8, 2]], Number = 2
 Output: 3

```python
n =[[1, 2, 3], [4, 2, 6], [7, 8, 2]]
count =0
number =2
for i in n:
    for j in i:
        if j ==number:
            count+=1
print("count:",count)        
    
```

Q4. Find and print the common elements between two lists.
 Input: List1 = [10, 20, 30, 40, 50], List2 = [30, 40, 60, 70, 80]
 Output: 30, 40


```python
List1 = [10, 20, 30, 40, 50]
List2 = [30, 40, 60, 70, 80]
for i in List1:
    for j in List2:
        if i ==j:
            print(i)
```

Q5. Merge two lists into one (without using + or extend).
 Input: List1 = [1, 2, 3], List2 = [4, 5, 6]
 Output: [1, 2, 3, 4, 5, 6]


```python
List1 = [1, 2, 3]
List2 = [4, 5, 6]
add =[]
for i in List1:
    add.append(i)
for j in List2:
    add.append(j)
print(add)    
```

Q6. Take 4 names from the user and store them in a list. Create a dictionary where:

Key → name


Value → dictionary with


"length" → length of the name


"vowels" → number of vowels in the name


Example Input: ["Ravi", "Anu"]
 Output:
{
 "Ravi": {"length": 4, "vowels": 2},
 "Anu": {"length": 3, "vowels": 2}
}






```python
name =["Ravi", "Anu"]
r ={}
for i in name:
    length =len(i)
    count =0
    for j in i: 
        if j in "aeiouAEIOU":
            count +=1
    r[i] = {"length":length,"vowels":count}
print(r)    

    
```

Q7. Take a sentence and create a new list where each word is reversed.
 Input: "hello world"
 Output: ["olleh", "dlrow"]


```python
n = "hello world"
r =[]
for i in n.split():
    r.append(i[::-1])
print(r)    
```
