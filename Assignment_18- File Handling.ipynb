Q3. Count how many times each vowel appears in the article.


```python
with open("article.txt","r")as file:
    content =file.read().lower()
    vowels="aeiou"
    for i in vowels:
        print(f"{i}: {content.count(i)}")
```

Q4. Find the longest word in the article.

```python
with open("article.txt","r")as file:
    content =file.read()
    words =content.split()
    longest = max(words,key =len)
    print(longest)
```

Q5. Print all sentences that contain a given keyword.


```python
keyword = input("Enter keyword: ")
with open("article.txt", "r") as file:
    content = file.read()
    sentences = content.split(".")
    for sentence in sentences:
        if keyword.lower() in sentence.lower():
            print(sentence.strip())
```

Q6. Count the number of words that start with a given letter.

```python
letter = input("Enter a letter: ")
with open("article.txt", "r") as file:
    content = file.read()
    words = content.split()
    count = 0
    for word in words:
        if word.lower().startswith(letter.lower()):
            count += 1
    print(f"starting with '{letter}':", count)
    
```

Q11. Use seek() and tell() to move the file pointer and display its position.

```python
with open("article.txt", "r") as file:
    print("Initial position:", file.tell())
    file.read(50)
    print("After reading 50 chars:", file.tell())
    file.seek(0)
    print("After seek(0):", file.tell())
```

Q14. Find how many unique words are in the article (ignore case).

```python
with open("article.txt", "r") as file:
    content = file.read().lower()
    words = content.split()
    unique_words = set(words)
    print("unique words:", len(unique_words))
```

Q15. Count how many times each word appears and save the result into word_count.txt.

```python
with open("article.txt", "r") as file:
    content = file.read().lower()
    words = content.split()
    word_count = {}
    for word in words:
        word_count[word] = word_count.get(word, 0) + 1
with open("word_count.txt", "w") as file:
    for word, count in word_count.items():
        file.write(f"{word}: {count}\n")

print("word_count.txt created successfully")
```
