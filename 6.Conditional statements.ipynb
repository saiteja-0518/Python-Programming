## Control statements 

- Controls the flow of program
- In python flow is handled by
 - Conditional statements 
   - If 
   - IF else
   - If elif
   - if elif else


# simple if

```python
m = int(input("enter the marks"))
if m>35:
    print("pass")
```

```python
# even or odd
n =int(input("enter a number:"))
if n%2==0:
    print("even")
else:
    print("odd")  
```

# if-else

```python
m = int(input("enter the marks"))
if m>35:
    print("pass")
else:
    print("fail")
```

## if-elif-else

```python
m = int(input("enter the marks"))
if m>=90:
    print("Grade A")
elif m>=70:      
    print("Grade B")
elif m >=50:
    print("Grade C")
else:
    print("fail")
```

## nested if

```python
age = int(input("enter the age"))
lie =input("enter either yes or no")
if age >=18:
    if lie.lower() =="yes":
        print("you can drive")
    else:
        print("can't drive")
else:
    print("not eligible")
```
