## Case Study – Banking System


Imagine you are working as a software developer for a bank. Your task is to create a banking application module with your name as the module name (e.g., priya_bank.py).
The module should handle:

Creating a new account


Crediting money


Debiting money


Handling invalid operations using custom exceptions


Ensuring that exceptions are properly managed with try-except-else-finally.



Requirements
1. Create a Python module
Name the file as <yourname>_bank.py (example: priya_bank.py).


Inside the module, create the following:


(a) Class: BankAccount
Attributes:
account_number (int)


account_holder (string)


balance (float, default = 0.0)


Methods:
create_account(account_number, account_holder, initial_balance=0.0)


credit(amount)


debit(amount) → raise an error if balance < amount


display_balance()

2. Exception Handling Rules
Use try-except to handle errors like:


Negative amount deposit


Negative withdrawal


Insufficient balance


Create a custom exception class:

 class InsufficientBalanceError(Exception):
    pass
Use else to confirm successful transactions.


Use finally to display "Transaction process completed".



3. Tasks (Demonstration File: bank_demo.py)
Write a separate file bank_demo.py and perform the following tasks by importing your module:
Task 1: Create a new account (Account number: 1001, Account holder: YourName, Initial balance: 5000). Print the account details.


Task 2: Try to credit 2000 into the account.


Use try-except-else-finally.


If successful, print "Credit successful".


If error, print the error message.


Always display "Transaction process completed".


Task 3: Try to debit 1000 from the account.


Handle exceptions properly.


If successful, print "Debit successful".


Else, print the error message.



Task 4: Try to debit 8000 from the account (more than available balance).


This should raise your InsufficientBalanceError.


Handle it using try-except.


Task 5: Try to credit a negative amount (-500).


This should raise a ValueError.


Handle it and print the error message.
