## Bitwise Operates 

```python
# AND ,OR Operators 
x = 2
y = 7
print('x & y:', x & y) 
print('x | y:', x | y)
```

```python
# XOR ,NOT Operators 
x = 2
y = 7
print('x ^ y:', x ^ y) 
print('~x :', ~x)
print('~y :', ~y)
```

```python
# left shift << ,right shift >>
x = 2
print('x <<1 :', x <<1)
print('x <<2 :', x <<2)
print('x <<3 :', x <<3)


y = 7
print('y <<4 :', y <<4)
print('y <<5 :', y <<5)
print('y << 6:', x <<6)



x = 2
print('x >>1 :', x >>1)
print('x >>2 :', x >>2)
print('x >>3 :', x >>3)


y = 7
print('y >>4 :', y >>4)
print('y >>5 :', y >>5)
print('y >> 6:', x >>6)
```
