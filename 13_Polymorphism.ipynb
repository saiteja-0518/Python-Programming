# Polymorphism

## What is Polymorphism?
- **Polymorphism** means **many forms**.
- The **same method name** behaves **differently** for **different classes**.

**Example:**  
A **Car** and a **Bike** both can `start()`, but the way they start is different.

## Why Use Polymorphism?
- **Flexibility:** Same method name, different behavior.
- **Cleaner code:** No need for too many conditions.
- **Easier to extend:** New vehicle types can be added easily.

## Method Overriding (Real Polymorphism in Python)
- A **child class** provides its **own version** of a method from the **parent class**.

**Example:**


```python
class Grandparent:
    def name(self):
        print("Grand parent name is pavan")
class Parent(Grandparent):
    def name(self):
        print("parent name is kumar")
class Child(Grandparent):
    def name (self):
        print("child name is sai")
names1 = Grandparent()
names1.name()

names2 = Parent()
names2.name()

names3 = Child()
names3.name()
```

```python
# Parent class
class Vehicle:
    def start(self):
        print("Vehicle is starting...")

# Child class 1
class Car(Vehicle):
    def start(self):
        print("Car engine starting with a button...")

# Child class 2
class Bike(Vehicle):
    def start(self):
        print("Bike starting with a key...")   
```

```python
vehicle1 = Vehicle()
vehicle1.start()
```

```python
car1 = Car()
# Overriding parent method with child method
car1.start()
```

```python
for vehicle in [Car(), Bike()]:
    vehicle.start()
```

# Polymorphism with Functions

- A function can work with different vehicle objects.

**Example:**



```python
def current_assent(Parent):
    Parent.append()
    Parent.UpperCase()
    Parent.name()
```

```python
def book_ride(vehicle):
    vehicle.start()
```

```python
book_ride(car1)
```

```python
book_ride(Car())
```

```python
# vehicle is a parameter with an expectation of class
def city_ride(vehicle):
    vehicle.start()
    
```

```python
city_ride(Car())
city_ride(Bike())
```

```python
def begin_journey(vehicle):
    vehicle.start()

begin_journey(Car())
begin_journey(Bike())
```

## Important Points
- Python supports polymorphism naturally through method overriding.
- True method overloading (same method name with different parameters) is not supported directly like in Java.
