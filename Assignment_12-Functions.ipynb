Q1. Write a function that takes a number as input and returns whether it is even or odd.
Input: 12

Output: "Even"


```python
def even_odd (n1):
    if n1%2==0:
        return("Even")
    else:
        print("Odd")
    return n1
print(even_odd(12))
```

Q4. Write a function that takes a list of numbers and returns the product of all numbers.
Input: [1, 3, 5]


Output: 15


```python
def product_nums(nums):
    p =1
    for i in nums:
        p *=i
    return p
print(product_nums([1,3,5]))
```

Q5. Write a function that takes a string and returns the number of vowels in it.
Input: "programming"


Output: 3


```python
def count_vowels(str):
    count =0
    vowels ="aeiouAEIOU" 
    for i in str:
        if i in vowels:
            count +=1
    return count
print(count_vowels("programming"))
```

Q9. Write a function that accepts a number and returns the factorial of that number.
Input: 5


Output: 120


```python
def factorial(num):
    fact =1
    for i in range(1,num+1):
        fact =fact*i
    return fact
print(factorial(5))
```

Q13. Write a function that takes a number and returns whether it is a prime number.
Input: 29


Output: True


```python
def prime_num(num):
    if num<=1:
        return False
    for i in range(2,int(num**0.5)+1):
        if num %i ==0:
            return False
    return True
print(prime_num(29))
    
```

Q14. Write a function that takes two numbers and returns their greatest common divisor (GCD).
Input: 36, 60


Output: 12


```python
def gcd(n1,n2):
    while n2 !=0:
        n1 ,n2 = n2, n1%n2
    return n1
print(gcd(36,60))
```

Q15. Write a function that takes a list of numbers and returns a new list with only even numbers.
Input: [1, 2, 3, 4, 5, 6]


Output: [2, 4, 6]


```python
def even_num(nums):
    return [i for i in nums if i % 2 == 0]
print(even_num([1, 2, 3, 4, 5, 6]))      
```

Q16. Write a function that takes a list of numbers and returns a new list with only odd numbers.
Input: [10, 11, 12, 13, 14]


Output: [11, 13]


```python
def odd_num(n):
    return [i for i in n if i%2!=0]
print(odd_num([10, 11, 12, 13, 14]))
```

Q18. Write a function that takes a string and returns the frequency of each character as a dictionary.
Input: "hello"


Output: {'h': 1, 'e': 1, 'l': 2, 'o': 1}


```python
def frequency(st):
    freq ={}
    for i in st:
        freq[i]= freq.get(i,0)+1
    return freq
print(frequency("hello"))
```

Q19. Write a function that accepts a number and returns a list of its multiplication table up to 10.
Input: 5


Output: [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]


```python
def table(n):
    return [n*i for i in range(1,11)]
print(table(5))
```

Q20. Write a function that takes two lists and returns a list of common elements.
Input: [1, 2, 3, 4], [3, 4, 5, 6]


Output: [3, 4]


```python
def common(l1,l2):
    return list(set(l1) & set(l2))
print(common([1, 2, 3, 4],[3, 4, 5, 6]))
```
